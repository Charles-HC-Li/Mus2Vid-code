```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio
import librosa
```

```python
y, sr = librosa.load("deam_dataset/DEAM_audio/MEMD_audio/3.mp3")
HOP_LENGTH = int(sr/2) + 1
y = y[15*sr:45*sr]
Audio(y, rate=sr)
```

```python
chroma_cq = librosa.feature.chroma_cqt(y=y, sr=sr, hop_length=HOP_LENGTH)
```

```python
np.shape(chroma_cq)
```

```python
fig, ax = plt.subplots(nrows=1, sharex=True, sharey=True)
img = librosa.display.specshow(chroma_cq, y_axis='chroma', x_axis='time',hop_length=HOP_LENGTH)
ax.set(title='chroma_cqt')
fig.colorbar(img, ax=ax)
```
